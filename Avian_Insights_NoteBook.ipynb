## Data Import

```python
import pandas as pd
```

```python
# Load Excel file 1
file_path = "Bird_Monitoring_Data_FOREST.xlsx"

# Read all sheets into a dictionary
all_sheets = pd.read_excel(file_path, sheet_name=None)

# Combine all sheets into one DataFrame
data_1 = pd.concat(all_sheets.values(), ignore_index=True)

# Load Excel file 2
file_path = "Bird_Monitoring_Data_GRASSLAND.xlsx"

# Read all sheets into a dictionary
all_sheets_2 = pd.read_excel(file_path, sheet_name=None)

# Combine all sheets into one DataFrame
data_2 = pd.concat(all_sheets_2.values(), ignore_index=True)

df = pd.concat([data_1,data_2], ignore_index=True)
print(df)
```

## Data Cleaning

```python
df.head()
df.info()
df.describe()
df.shape
```

```python
df.isna().sum()
```

```python
df = df.dropna(how = 'all')
df.NPSTaxonCode = df.NPSTaxonCode.astype(str)
df.AcceptedTSN = df.AcceptedTSN.astype(str)
```

```python
df.NPSTaxonCode.replace({'nan' : 'Unknown'}, inplace=True)
df.AcceptedTSN.replace({'nan' : 'Unknown'}, inplace=True)
df.Sex.replace({'Undetermined' : 'Unknown'}, inplace=True)
```

```python
df.Sub_Unit_Code.fillna('Unknown',inplace=True)
df.Site_Name.fillna('Unknown',inplace=True)
df.ID_Method.fillna(df.ID_Method.mode()[0],inplace=True)
df.Distance.fillna(df.Distance.mode()[0],inplace=True)
df.Site_Name.fillna('Unknown', inplace=True)
df.Sex.fillna('Unknown',inplace=True)
df = df.drop(columns = 'Previously_Obs')
```

```python
distance_update = {'<= 50 Meters' : 25,
'50 - 100 Meters' : 75
}
df.Distance = df.Distance.map(distance_update).astype(int)
```

```python
interval_length_update = {'0-2.5 min' : 1.25,
    '2.5 - 5 min' : 3.75,
'5 - 7.5 min' : 6.25,
'7.5 - 10 min' : 8.75
}
df.Interval_Length = df.Interval_Length.map(interval_length_update).astype(float)
```

```python
df.isna().sum()
```

```python
df.Temperature = df.Temperature.astype(int)
df.Humidity   = df.Humidity.astype(int)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

## Exploratory Data Analysis

```python
# 1. Temporal Analysis
# Seasonal Trends
df['Month'] = df['Date'].dt.month
# year trend
yearly = df.groupby('Year')['Scientific_Name'].count()
print(yearly)
# month trend
monthly = df.groupby('Month')['Scientific_Name'].count()
print(monthly)

# Observation Time Analysis
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='%H:%M:%S')
df['End_Time'] = pd.to_datetime(df['End_Time'], format='%H:%M:%S')

time_activity = df.groupby(df['Start_Time'].dt.hour)['Scientific_Name'].count()
print(time_activity)
```

```python
# 2. Spatial Analysis
# Location Insights
location_group = df.groupby('Location_Type')['Scientific_Name'].nunique()
print(location_group)

# Plot-Level Analysis
plot_analysis = df.groupby('Plot_Name')['Scientific_Name'].count()
print(plot_analysis)
```

```python
# 3. Species Analysis
# Diversity Metrics
total_species = df['Scientific_Name'].nunique()
print('total_species :',total_species)
species_distribution = df.groupby('Location_Type')['Scientific_Name'].nunique()
print(species_distribution)

# Activity Patterns
activity = df.groupby(['Interval_Length', 'ID_Method']).size()
print(activity)

# Sex Ratio
sex_ratio = df.groupby(['Scientific_Name', 'Sex']).size().unstack(fill_value=0)
print(sex_ratio)
```

```python
# 4. Environmental Conditions
# Weather Correlation
weather = df.groupby(['Sky', 'Wind'])['Scientific_Name'].size()
temp_effect = df.groupby('Temperature')['Scientific_Name'].count()
humidity_effect = df.groupby('Humidity')['Scientific_Name'].count()

# Disturbance Effect
disturbance = df.groupby('Disturbance')['Scientific_Name'].count()
```

```python
# 5. Distance and Behaviour
# Distance Analysis
distance_species = df.groupby('Distance')['Scientific_Name'].value_counts()
print(distance_species)

# Flyover Frequency
flyover = df.groupby('Flyover_Observed')['Scientific_Name'].value_counts()
print(flyover)
```

```python
# 6. Observer Trends
# Observer Bias
observer = df.groupby('Observer')['Scientific_Name'].count()
print(observer)

# Visit Patterns
visit = df.groupby('Visit')['Scientific_Name'].nunique()
print(visit)
```

```python
# 7. Conservation Insights
# Watchlist Trends
watchlist = df.groupby('PIF_Watchlist_Status')['Scientific_Name'].count()
stewardship = df.groupby('Regional_Stewardship_Status')['Scientific_Name'].count()
print(watchlist)
print(stewardship)

# AOU Code Patterns
aou = df.groupby('AOU_Code')['Scientific_Name'].count()
print(aou)
```
